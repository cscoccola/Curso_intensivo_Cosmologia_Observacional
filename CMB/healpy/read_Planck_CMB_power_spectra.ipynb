## Generate a CMB map realization

The power spectrum gives us the distribution of power with the angular scale, if we want to create a map we could use hp.synfast, however the realization will be different each time that we run synfast, we could always use the same seed to make it reproducible. However, in case we want to have the same realization of a CMB map at different resolutions ($N_{side}$), it is better to generate a realization of the spectrum in Spherical Harmonics space, creating a set of $a_{\ell m}$, and then when needed transform them to a map with hp.alm2map.

We also want to make sure that the $a_{\ell m}$ have the same $\ell_{max}$ of the input $C_\ell$:



```python
data_path = "/home/claudia/Dropbox/CLAUDIA/TRABAJO/DOCENCIA/curso_intensivo_Cosmologia/data/"

spectra_file =  "COM_PowerSpect_CMB-base-plikHM-TTTEEE-lowl-lowE-lensing-minimum-theory_R3.01.txt"

full_path_spectra_file = data_path + spectra_file
```

```python

# Cargar el espectro, ignorando las líneas comentadas
spectra = np.loadtxt(full_path_spectra_file)

# Verificar los primeros datos
print(spectra[:5])  # Muestra las primeras 5 filas para comprobar si está bien cargado
```

## CORREGIR A PARTIR DE ACA PORQUE ESTA MAL

```python
# Transponer el array para que las columnas sean filas
spectra_transposed = spectra.T

# Verificar la transposición
print(spectra_transposed[:5])  # Muestra las primeras 5 filas de los espectros transpuestos

# Ahora, cada fila será un espectro:
l = spectra_transposed[0]  # La primera fila (L)
tt = spectra_transposed[1]  # La segunda fila (TT)
te = spectra_transposed[2]  # La tercera fila (TE)
ee = spectra_transposed[3]  # La cuarta fila (EE)
bb = spectra_transposed[4]  # La quinta fila (BB)

# Verificar los primeros datos de cada espectro transpuesto
print(f"L: {l[:5]}")
print(f"TT: {tt[:5]}")
print(f"TE: {te[:5]}")
print(f"EE: {ee[:5]}")
print(f"BB: {bb[:5]}")
```
